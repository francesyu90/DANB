```python
df_popularity_genres['genres'].str.contains('|')
```

```python
hb_1 = df_popularity_genres[df_popularity_genres['genres'].str.contains('|')]
```

```python
import numpy as np
```

```python
num_genres = []
genres = []

for index, row in hb_1.iterrows():
    splitted_genres = row.get('genres').split('|')
    num_genres.append(len(splitted_genres))
    genres.extend(splitted_genres)

target_genres = list(set(genres))
```

```python


for index, row in df_popularity_genres.iterrows():
    
#     print(row.get('genres'))
    
    splitted_genres = row.get('genres').split('|')
    
    size = len(splitted_genres)
    
    for i in range(size):
        df_popularity_genres[splitted_genres[i]] = True
        

#     df_popularity_genres.drop(index, inplace=True)
    
#     df_popularity_genres.append(tmp_df, ignore_index=True)
    
print(df_popularity_genres)
    
```
